## 📖 문자열(String)이란 무엇인가?

### 문자열의 정의

문자열은 **텍스트 데이터를 저장하는 자료형**입니다. 이름, 주소, 이메일, 메시지 등 우리가 다루는 대부분의 텍스트 정보가 문자열로 표현됩니다.

```python
name = "홍길동"          # 이름
email = "hong@example.com"  # 이메일
message = "안녕하세요!"    # 메시지
```

### 왜 문자열이 중요할까요?

**1. 웹 개발의 핵심**
- 사용자 이름, 이메일, 비밀번호 등 대부분의 입력이 문자열
- API 응답도 대부분 문자열(JSON)로 전달

**2. FastAPI에서의 활용**
```python
# FastAPI 엔드포인트에서 사용자 정보를 문자열로 처리
@app.get("/users/{user_id}")
def get_user(user_id: str):  # ← 문자열!
    return {"name": "홍길동"}  # ← 문자열!
```

**3. 데이터 검증과 가공**
- 이메일 형식 체크
- 비밀번호 길이 확인
- 사용자 입력 정제

### 문자열의 불변성(Immutability)

> ⚠️ **중요**: 문자열은 한번 만들어지면 **변경할 수 없습니다**. 이것을 "불변성"이라고 합니다.

```python
name = "Alice"
# name의 첫 글자를 바꾸려고 하면 에러!
# name[0] = "B"  # 이렇게 하면 TypeError 발생

# 대신 새로운 문자열을 만들어야 합니다
name = "Bob"  # OK!
```

불변성 때문에 문자열은 **안전**하고 **예측 가능**합니다. 특히 멀티스레드 환경에서 중요한 특성입니다!

## 1. 문자열 생성 - 작은따옴표 vs 큰따옴표

### 두 가지 방법, 하나의 결과

Python에서 문자열을 만드는 방법은 두 가지입니다:

1. **작은따옴표** `'문자열'`
2. **큰따옴표** `"문자열"`

두 방법은 **완전히 동일**합니다! 어떤 것을 사용해도 똑같은 문자열이 만들어집니다.

### 💡 언제 어떤 따옴표를 쓸까요?

**일반적인 경우**: 아무거나 사용하세요! 팀 컨벤션에 따르면 됩니다.

**문자열 안에 따옴표가 포함된 경우**: 서로 다른 따옴표를 사용하세요.
```python
# 작은따옴표 안에 큰따옴표
message1 = 'He said, "Hello!"'  # OK

# 큰따옴표 안에 작은따옴표  
message2 = "She said, 'Hi!'"  # OK
```

### 실무 팁

- **FastAPI**: 큰따옴표를 주로 사용 (JSON이 큰따옴표를 사용하므로)
- **일관성**: 프로젝트 내에서 하나로 통일하는 것이 좋습니다
- **에디터 설정**: VS Code에서 자동 포매팅 설정 가능

````python
# 작은따옴표로 문자열 생성
name1 = 'Alice'

# 큰따옴표로 문자열 생성
name2 = "Bob"

# 둘 다 똑같은 문자열 타입입니다
print(name1)  # 출력: Alice
print(name2)  # 출력: Bob
print(type(name1))  # 출력: <class 'str'>
print(type(name2))  # 출력: <class 'str'>
````

**실행 결과:**
```
Alice
Bob
<class 'str'>
<class 'str'>
```

두 변수 모두 `str` (string) 타입입니다. 어떤 따옴표를 사용했는지와 관계없이 동일한 타입이죠!

````python
# 문자열 안에 따옴표를 포함하는 방법

# 방법 1: 큰따옴표 안에 작은따옴표
message1 = "He said, 'Hello!'"
print(message1)  # 출력: He said, 'Hello!'

# 방법 2: 작은따옴표 안에 큰따옴표
message2 = 'She said, "Hi!"'
print(message2)  # 출력: She said, "Hi!"

# ⚠️ 같은 따옴표를 중첩하면 에러!
# wrong = "He said, "Hello!""  # SyntaxError!
````

**실행 결과:**
```
He said, 'Hello!'
She said, "Hi!"
```

### ✅ 핵심 정리
- 작은따옴표와 큰따옴표는 **기능상 완전히 동일**
- 문자열 내부에 따옴표가 필요하면 **서로 다른 따옴표 사용**
- 같은 따옴표를 중첩하면 문법 에러 발생

## 2. 문자열 연산 - 덧셈(+)으로 합치기

### 문자열 연결(Concatenation)

숫자는 `+`로 더하지만, 문자열은 `+`로 **연결**됩니다! 이것을 "문자열 연결" 또는 "concatenation"이라고 합니다.

```python
# 숫자의 덧셈
1 + 2  # → 3

# 문자열의 연결
"Hello" + "World"  # → "HelloWorld"
```

### 💡 실무에서 언제 사용할까요?

**1. 사용자 정보 조합**
```python
first_name = "길동"
last_name = "홍"
full_name = last_name + first_name  # "홍길동"
```

**2. API 경로 생성**
```python
base_url = "https://api.example.com"
endpoint = "/users"
full_url = base_url + endpoint  # "https://api.example.com/users"
```

**3. 메시지 조합**
```python
username = "Alice"
greeting = "안녕하세요, " + username + "님!"  # "안녕하세요, Alice님!"
```

### ⚠️ 주의사항: 공백 처리

문자열을 연결할 때 **공백은 자동으로 추가되지 않습니다**! 필요하면 직접 넣어야 합니다.

```python
# 공백 없이 연결
"Hello" + "World"  # "HelloWorld" (붙어있음!)

# 공백 포함해서 연결
"Hello" + " " + "World"  # "Hello World" (공백 있음!)
```

````python
# 기본 문자열 연결
first_name = "홍"
last_name = "길동"

# + 연산자로 두 문자열을 연결
full_name = first_name + last_name

print(full_name)  # 출력: 홍길동
print(type(full_name))  # 출력: <class 'str'>
````

**실행 결과:**
```
홍길동
<class 'str'>
```

두 개의 문자열이 하나로 합쳐졌습니다! 결과도 문자열 타입이죠.

````python
# 여러 문자열을 연결해서 문장 만들기

# 사용자 이름
name = "홍길동"

# 여러 부분을 + 연산자로 연결
greeting = "안녕하세요, " + name + "님!"

print(greeting)  # 출력: 안녕하세요, 홍길동님!

# 주의: 공백이 없으면 어떻게 될까요?
greeting_no_space = "안녕하세요," + name + "님!"
print(greeting_no_space)  # 출력: 안녕하세요,홍길동님! (공백 없음!)
````

**실행 결과:**
```
안녕하세요, 홍길동님!
안녕하세요,홍길동님!
```

### 📌 중요 포인트
- 공백을 포함하려면 `" "` 문자열을 명시적으로 추가해야 합니다
- `+` 연산자는 여러 번 사용할 수 있습니다
- 나중에 배울 f-string이 더 편리하지만, 기본 연결 방법도 알아두세요!

## 3. 문자열 연산 - 곱셈(*)으로 반복하기

### 문자열 반복(Repetition)

문자열에 숫자를 곱하면 그 문자열이 **반복**됩니다! 매우 편리한 기능이죠.

```python
"Ha" * 3  # → "HaHaHa"
"=" * 10  # → "=========="
```

### 💡 실무에서 언제 사용할까요?

**1. 구분선 만들기**
```python
print("=" * 50)  # 예쁜 구분선
print("제목")
print("=" * 50)
```

**2. 로딩 인디케이터**
```python
loading = "." * 3  # "..."
print(f"처리 중{loading}")
```

**3. 패턴 생성**
```python
pattern = "*-" * 10  # "*-*-*-*-*-*-*-*-*-*-"
```

**4. 테스트 데이터**
```python
dummy_text = "test" * 100  # 긴 텍스트 생성
```

### ⚠️ 주의사항

- 곱하는 숫자가 **0 이하**면 **빈 문자열**이 됩니다
- 너무 큰 숫자를 곱하면 메모리 문제가 발생할 수 있습니다

````python
# 한 글자 반복하기
star = "*"
line = star * 10  # "*"를 10번 반복

print(line)  # 출력: **********
print(len(line))  # 출력: 10 (10개의 별표)
````

**실행 결과:**
```
**********
10
```

한 글자 `*`가 10번 반복되어 10글자 문자열이 되었습니다!

````python
# 여러 글자도 반복 가능!
pattern = "=="
repeated_pattern = pattern * 5  # "=="를 5번 반복

print(repeated_pattern)  # 출력: ==========
print(len(repeated_pattern))  # 출력: 10 (2글자 × 5번 = 10글자)

# 다양한 패턴 만들기
separator1 = "-" * 20
separator2 = "~" * 20
separator3 = "*-" * 10

print(separator1)  # 출력: --------------------
print(separator2)  # 출력: ~~~~~~~~~~~~~~~~~~~~
print(separator3)  # 출력: *-*-*-*-*-*-*-*-*-*-
````

**실행 결과:**
```
==========
10
--------------------
~~~~~~~~~~~~~~~~~~~~
*-*-*-*-*-*-*-*-*-*-
```

### 실전 활용: 보기 좋은 출력 만들기

````python
# 제목과 구분선으로 예쁘게 출력하기

title = "Python 문자열 학습"
separator = "=" * 30  # 30개의 = 기호

# 예쁘게 출력
print(separator)
print(title)
print(separator)
print()
print("내용이 여기에 들어갑니다...")
print()
print(separator)
````

**실행 결과:**
```
==============================
Python 문자열 학습
==============================

내용이 여기에 들어갑니다...

==============================
```

### ✅ 핵심 정리
- `문자열 * 숫자`로 문자열을 반복할 수 있습니다
- 한 글자든 여러 글자든 모두 반복 가능합니다
- 구분선, 패턴, 로딩 표시 등 다양하게 활용됩니다

## 4. len() 함수 - 문자열 길이 구하기

### len() 함수란?

`len()`은 **length(길이)**의 줄임말입니다. 문자열의 **글자 수**를 반환하는 내장 함수입니다.

```python
len("Hello")  # → 5 (5글자)
len("안녕")    # → 2 (2글자)
```

### 💡 실무에서 언제 사용할까요?

**1. 입력 검증**
```python
password = "abc123"
if len(password) < 8:
    print("비밀번호는 8자 이상이어야 합니다!")
```

**2. 데이터 제한**
```python
tweet = "트위터 메시지"
if len(tweet) > 280:
    print("280자를 초과했습니다!")
```

**3. 사용자 피드백**
```python
message = "안녕하세요"
remaining = 100 - len(message)
print(f"남은 글자: {remaining}")
```

**4. FastAPI에서의 활용**
```python
from pydantic import BaseModel, validator

class User(BaseModel):
    username: str
    
    @validator('username')
    def username_length(cls, v):
        if len(v) < 3:  # ← len() 사용!
            raise ValueError('Username too short')
        return v
```

### ⚠️ 주의사항

**1. 공백도 글자 수에 포함됩니다**
```python
len("Hello World")  # → 11 (공백 포함!)
```

**2. 한글도 영어와 똑같이 1글자 = 1**
```python
len("안녕")  # → 2
len("AB")   # → 2 (똑같음!)
```

**3. 이모지는 글자 수가 다를 수 있습니다**
```python
len("😀")  # → 1 (보통은 1이지만 환경에 따라 다를 수 있음)
```

````python
# 기본 사용법
message = "Hello"
length = len(message)  # 문자열의 길이를 계산

print(message)  # 출력: Hello
print(length)  # 출력: 5
print(type(length))  # 출력: <class 'int'> (정수 타입!)
````

**실행 결과:**
```
Hello
5
<class 'int'>
```

`len()` 함수는 **정수(int)**를 반환합니다!

````python
# 한글 문자열의 길이
korean = "안녕하세요"
print(f"문자열: {korean}")
print(f"길이: {len(korean)}글자")  # 출력: 길이: 5글자

# 영어와 동일하게 계산됩니다
english = "Hello"
print(f"\n문자열: {english}")
print(f"길이: {len(english)}글자")  # 출력: 길이: 5글자

# 둘 다 5글자!
````

**실행 결과:**
```
문자열: 안녕하세요
길이: 5글자

문자열: Hello
길이: 5글자
```

한글이든 영어든 **1글자는 1**로 계산됩니다!

````python
# 공백도 글자 수에 포함됩니다!
sentence = "Hello World"
print(sentence)
print(f"길이: {len(sentence)}글자")  # 출력: 길이: 11글자

# 공백 없이 연결하면?
no_space = "HelloWorld"
print(no_space)
print(f"길이: {len(no_space)}글자")  # 출력: 길이: 10글자

# 공백 1개 차이!
````

**실행 결과:**
```
Hello World
길이: 11글자
HelloWorld
길이: 10글자
```

### 실전 활용: 입력 검증

````python
# 비밀번호 길이 검증 시뮬레이션
password1 = "abc123"  # 너무 짧음
password2 = "abcd1234"  # 적당함

# 최소 길이 기준
min_length = 8

# 첫 번째 비밀번호 체크
print(f"비밀번호: {password1}")
print(f"길이: {len(password1)}자")
if len(password1) < min_length:
    print(f"❌ 비밀번호는 {min_length}자 이상이어야 합니다!")
else:
    print("✅ 비밀번호 길이가 적절합니다.")

print()  # 줄바꿈

# 두 번째 비밀번호 체크
print(f"비밀번호: {password2}")
print(f"길이: {len(password2)}자")
if len(password2) < min_length:
    print(f"❌ 비밀번호는 {min_length}자 이상이어야 합니다!")
else:
    print("✅ 비밀번호 길이가 적절합니다.")
````

**실행 결과:**
```
비밀번호: abc123
길이: 6자
❌ 비밀번호는 8자 이상이어야 합니다!

비밀번호: abcd1234
길이: 8자
✅ 비밀번호 길이가 적절합니다.
```

### ✅ 핵심 정리
- `len()` 함수는 문자열의 글자 수를 **정수**로 반환합니다
- 공백, 특수문자 모두 글자 수에 포함됩니다
- 한글과 영어는 동일하게 1글자 = 1로 계산됩니다
- 입력 검증, 데이터 제한 등 실무에서 필수적으로 사용됩니다

### 종합 활용 예제

지금까지 배운 내용을 모두 활용해서 **쇼핑몰 상품 정보를 표시**하는 프로그램을 만들어봅시다!

**사용할 기술:**
- 문자열 생성 (따옴표)
- 문자열 연결 (+)
- 문자열 반복 (*)
- len() 함수
- f-string (미리 맛보기!)

### 💡 실무 시나리오

쇼핑몰 API에서 상품 정보를 받아서 사용자에게 보기 좋게 표시하는 상황입니다.

```python
# FastAPI에서 이렇게 사용될 수 있습니다
@app.get("/products/{product_id}")
def get_product(product_id: int):
    # 상품 정보 조합
    brand = "삼성"
    model = "갤럭시북"
    product_name = "노트북"
    
    full_product = brand + " " + model + " " + product_name
    
    return {"product": full_product}
```

````python
# 상품 정보 변수 설정
product_name = "노트북"
brand = "삼성"
model = "갤럭시북"

# 문자열 연산으로 상품 정보 조합
# + 연산자로 여러 문자열을 연결하고, 공백도 추가
full_product = brand + " " + model + " " + product_name

print(full_product)  # 출력: 삼성 갤럭시북 노트북
````

**실행 결과:**
```
삼성 갤럭시북 노트북
```

세 개의 정보가 공백과 함께 하나의 문자열로 조합되었습니다!

````python
# 구분선으로 보기 좋게 표시하기

# * 연산자로 30개의 = 기호 만들기
separator = "=" * 30

# 예쁘게 출력
print(separator)  # 위쪽 구분선
print(full_product)  # 상품 정보
print(separator)  # 아래쪽 구분선
````

**실행 결과:**
```
==============================
삼성 갤럭시북 노트북
==============================
```

구분선으로 훨씬 보기 좋아졌네요!

````python
# 상품명 길이 확인하기

# len() 함수로 문자열 길이 계산
name_length = len(full_product)

# f-string으로 예쁘게 출력 (Day 1-6교시에서 자세히 배웁니다!)
print(f"상품명: {full_product}")
print(f"상품명 길이: {name_length}글자")

# 각 부분의 길이도 확인
print(f"\n[상세 정보]")
print(f"브랜드: {brand} ({len(brand)}글자)")
print(f"모델: {model} ({len(model)}글자)")
print(f"제품명: {product_name} ({len(product_name)}글자)")
print(f"공백: 2글자")
print(f"전체: {name_length}글자")
````

### 오늘 배운 내용

1. **문자열 생성**: 작은따옴표(`'`)와 큰따옴표(`"`) 모두 사용 가능
   - 기능은 완전히 동일
   - 문자열 안에 따옴표가 필요하면 서로 다른 종류 사용

2. **문자열 연결(+)**: 여러 문자열을 하나로 합치기
   - `"Hello" + " " + "World"` → `"Hello World"`
   - 공백은 자동으로 추가되지 않음 (직접 넣어야 함)

3. **문자열 반복(*)**: 문자열을 여러 번 반복
   - `"*" * 10` → `"**********"`
   - 구분선, 패턴, 로딩 표시 등에 활용

4. **len() 함수**: 문자열의 길이(글자 수) 계산
   - `len("Hello")` → `5`
   - 공백, 특수문자 모두 포함
   - 한글도 1글자 = 1로 계산
